```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("/content/AI_Impact_on_Jobs_2030.csv")
df.head(50)
```

```python
data = df.iloc[:, :7]
data.head(50)
```

```python
conexao = sqlite3.connect('dados.db')
data.to_sql('ai_impact_data', conexao, if_exists='replace', index=False)
print("Dados salvos com sucesso!!")
verificar_df = pd.read_sql_query("SELECT * FROM ai_impact_data LIMIT 5", conexao)
display(verificar_df)

conexao.close()
```
